```python
import os
from matplotlib import pyplot as plt
import tensorflow as tf 
import tensorflow_io as tfio
```

```python
#test
wav1 = os.path.join('data', 'Parsed_Capuchinbird_Clips', 'XC3776-1.wav')
print(wav1)
wav1_content = tf.io.read_file(wav1)
print(wav1_content)
wav1, sample_rate = tf.audio.decode_wav(wav1_content, desired_channels=1)
print(wav1)  
print (sample_rate) 

print("transform")
wav1 = tf.squeeze(wav1, axis = -1)
sample_rate = tf.cast(sample_rate, dtype=tf.int64) #float32 -> float 64
print(wav1)
print(wav1.shape)
print(sample_rate)

print("transform 2")
wav1 = wav1.numpy()
print(wav1)

```
